```python
import pandas as pd
import numpy as np

# Assume 'df' is your DataFrame
column_name = 'temperature_drop_celsius'

# 1. (Optional) See the data type before cleaning
print(f"--- Data type BEFORE cleaning ---")
print(new[column_name].dtype)


# 2. Clean the column (Robust 2-Step Method)
# First, ensure it's a string
df[column_name] = df[column_name].astype(str)

# Step 2a: Replace any commas with decimal points
# This turns 'q,2' into 'q.2' and '14,2' into '14.2'
df[column_name] = df[column_name].str.replace(',', '.', regex=False)

# Step 2b: Now, remove all remaining non-numeric characters
# This turns 'q.2' into '.2' and '14.z' into '14.'
df[column_name] = df[column_name].str.replace(r'[^0-9.]', '', regex=True)


# 3. Convert the cleaned column to a numeric type (float)
# 'coerce' will turn any remaining bad values (e.g., empty strings) into NaN
# '.2' becomes 0.2
# '14.2' becomes 14.2
# '14.' becomes 14.0
df[column_name] = pd.to_numeric(df[column_name], errors='coerce')

# 4. (Optional) Check the column after cleaning
print(f"--- Data type AFTER cleaning ---")
print(df[column_name].dtype)
print(df[column_name].head())
```
